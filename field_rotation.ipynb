```python
import numpy as np
import matplotlib.pyplot as plt
# Constants
latitude = np.radians(54 + 41/60) # latitude in radians (e.g., 54 degrees 41 minutes)
azimuths = np.linspace(0, 2 * np.pi, 360) # azimuths from 0 to 360 degrees (in radians)
altitudes = np.linspace(0, np.pi / 2, 180) # altitudes from 0 to 90 degrees (in radians)
# Function to calculate maximum exposure time
def max_exposure_time(altitude, azimuth):
    return np.abs(12.24 * np.cos(altitude) / (np.cos(latitude) * np.cos(azimuth)))
# Create a meshgrid of azimuths and altitudes
Azi, Alt = np.meshgrid(azimuths, altitudes)
# Calculate the maximum exposure values for each pair (azimuth, altitude)
MaxExposure = max_exposure_time(Alt, Azi)
# Create a color map: >30 green, 20-30 yellow, 10-20 orange, <10 red
cmap = plt.get_cmap('RdYlGn')
levels = [0, 10, 20, 30, np.max(MaxExposure)]
colors = ['red', 'orange', 'yellow', 'green']
# Plotting the graph
plt.figure(figsize=(20, 8))
contour = plt.contourf(Azi * 180 / np.pi, Alt * 180 / np.pi, MaxExposure, levels=levels, colors=colors)
# Adding a color bar
cbar = plt.colorbar(contour)
cbar.set_label('Max Exposure Time')
# Setting up the axes
plt.xlabel('Azimuth (degrees)')
plt.ylabel('Altitude (degrees)')
plt.title(f'Max Exposure Time for Seestar S50 field of view on latitude {latitude * 180 / np.pi:.1f}')
plt.grid(True)
# Displaying the graph
plt.xticks(np.arange(0, 361, 10), np.arange(0, 361, 10)) # X-axis labels every 10 degrees
plt.grid(True, which='both', axis='x') # Enable grid on the X-axis
plt.show()
```
